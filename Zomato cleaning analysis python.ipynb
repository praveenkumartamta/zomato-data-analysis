# Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('zomato dataset.csv')
```

```python
#Total duplicate
data.duplicated(subset=['Restaurant Name', 'Address', 'Phone Number']).sum()
```

```python
#Show duplicate
data[data.duplicated(subset=['Restaurant Name', 'Address', 'Phone Number'], keep=False)]
```

```python
#Remove duplicate
Remove_duplicate = data.drop_duplicates(subset=['Restaurant Name', 'Address', 'Phone Number'], keep='first',inplace=True )
```

```python
data.info()
```

```python
#Replace unknown with NAN
data['Average Cost for Two'] = data['Average Cost for Two'].replace('unknown',np.NAN)
```

```python
#Changed the data type from str to float
data['Average Cost for Two'] = data['Average Cost for Two'].astype(float)
```

```python
#Replaced null values with unknown
data['Cuisines'] = data['Cuisines'].fillna(value = 'Unknown')
```

```python
#Replaced null values with the average (mean)
data['Average Cost for Two'] = data['Average Cost for Two'].fillna(value = data['Average Cost for Two'].mean())
```

```python
#Replaced null values with the average (mean)
data['Aggregate Rating'] = data['Aggregate Rating'].fillna(value = data['Aggregate Rating'].mean())
```

```python
#Replaced null values with the average (mean)
data['Votes'] = data['Votes'].fillna(value = data['Votes'].mean())
```
